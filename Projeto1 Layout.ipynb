```python
#Funcao que transforma a renda (quantitativa) para ser analisada como qualitativa
def faixasph(df):
    x=df['V4621']
    if x<600:
        return 1
    elif x<1200:
        return 2
    elif x<1800:
        return 3
    elif x<2400:
        return 4
    elif x<3000:
        return 5
    elif x<3600:
        return 6
    elif x<4200:
        return 7
    elif x<4800:
        return 8
    elif x<5400:
        return 9
    elif x<6000:
        return 10
    return 11
faixas=dados.apply(faixasph,axis=1)
faixasN=dadosN.apply(faixasph,axis=1)
faixasN2=dadosN2.apply(faixasph,axis=1)
dados['faixas']=faixas
dadosN['faixas']=faixas
dadosN2['faixas']=faixas
dados.faixas=dados.faixas.astype('category')
dadosN.faixas=dadosN.faixas.astype('category')
dadosN2.faixas=dadosN2.faixas.astype('category')
dados.faixas.cat.categories=(['600','1200','1800','2400','3000','3600','4200','4800','5400','6000','+6000'])
dadosN.faixas.cat.categories=(['600','1200','1800','2400','3000','3600','4200','4800','5400','6000','+6000'])
dadosN2.faixas.cat.categories=(['600','1200','1800','2400','3000','3600','4200','4800','5400','6000','+6000'])
```

Redes de Streaming podem ser utilizadas por inúmeros meios e o acesso à internet atualmente não é algo tão caro, facilitando para as faixas de renda mais baixa, sendo um bom meio de obter muito conteúdo por bem menos preço. Apesar de tudo, o que mantém as redes de TV por assinatura ainda vivas, são domicílios com renda altíssima com a compra de planos caros, apesar disso, analisando a passagem dos anos, cada vez menos domicílios optam por manter ou comprar uma TV por assinatura e preferem bem mais manter uma rede de streaming. Logo, pode-se concluir que a queda da TV por assinatura se deve em grande parte à ascensão das redes de streaming, sendo a última um melhor custo/benefício, principalmente pensando nas classes baixas e médias.
